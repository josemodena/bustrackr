```python
import requests
url = 'http://bustime.mta.info/api/siri/vehicle-monitoring.json'
payload = {'key': '740c5f9a-c180-42e0-bafc-5723516152bf',
           'version': '2',
           'VehicleMonitoringDetailLevel': 'minimum'}
mta_response = requests.get(url, params=payload)
```

```python
mta_json = mta_response.json()
mta_json
```

```python
vehicles_info = mta_json['Siri']['ServiceDelivery']['VehicleMonitoringDelivery'][0]['VehicleActivity']
type(vehicles_info)
```

```python
print 'There are %s buses on the road now' % len(vehicles_info)
```

```python
vehicle_destination = vehicles_info[vehicle_number]['MonitoredVehicleJourney']['DestinationName']
vehicle_destination
```

```python
popup_text = 'Destination: %s' % vehicle_destination
popup_text
```

```python
list_vehicles = range(0, 20)

location = 'New York City, NY, US'
loc = geocoder.google(location)

map1 = folium.Map(location=loc.latlng, zoom_start=10)

for vehicle_number in list_vehicles:
    
    vehicle_latlng = [vehicles_info[vehicle_number]['MonitoredVehicleJourney']['VehicleLocation']['Latitude'], vehicles_info[vehicle_number]['MonitoredVehicleJourney']['VehicleLocation']['Longitude']]    
    vehicle_destination = vehicles_info[vehicle_number]['MonitoredVehicleJourney']['DestinationName']
    popup_text = 'Destination: %s' % vehicle_destination

    folium.Marker(vehicle_latlng, popup=popup_text).add_to(map1)

map1
```
